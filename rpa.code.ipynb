```python
import pandas as pd
import os
```

```python
df=pd.read_csv("bmw_global_sales_2018_2025.csv")
```

```python
df.head(3
        )
```

```python
df.info()
```

```python
#We don't have any gaps, so our data editing work is finished
```

```python
#Now let's find out the total sales figures by region and year.
```

```python
region_year_sales=df.groupby(["Region","Year"])["Units_Sold"].sum().reset_index()
print("Sales by region and year",region_year_sales.head())
```

```python
model_sales=df.groupby(["Model"])["Units_Sold"].sum().sort_values(ascending=False).reset_index()
print("Sales by model",model_sales.head())
```

```python
bev_premium_summary = df.groupby("Region")[["BEV_Share", "Premium_Share"]].mean().reset_index()
print("BEV and Premium Segment Averages:",bev_premium_summary.head())
```

```python
#Let's create a report for Excel.
```

```python
output_folder="reports"
os.makedirs(output_folder,exist_ok=True)
```

```python
region_year_sales.to_excel(os.path.join(output_folder, "region_year_sales.xlsx"), index=False)
model_sales.to_excel(os.path.join(output_folder, "model_sales.xlsx"), index=False)
bev_premium_summary.to_excel(os.path.join(output_folder, "ev_premium_summary.xlsx"), index=False)
```
